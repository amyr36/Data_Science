```python
import numpy as np
import pandas as pd
```

```python
students_grades = np.random.random_integers(7, 20, size=(15, 5))
studants_numbers = np.arange(1, 16)
```

```python
greades_of_term = pd.DataFrame(students_grades,
                               index=[studants_numbers],
                               columns=["math", "physics", "chemistry", "social sciences", "literature"])
```

```python
students_grades_average = np.average(students_grades, weights=[4, 4, 3, 3, 2], axis=1)
students_grades_average
```

```python
greades_of_term["average"] = students_grades_average.T
```

```python
students_grades_var = np.var(students_grades, axis=1)
students_grades_var
```

```python
greades_of_term["var"] = students_grades_var.T
greades_of_term
```
