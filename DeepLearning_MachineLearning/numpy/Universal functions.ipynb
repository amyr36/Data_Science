```python
import numpy as np
```

```python
def my_func(x):
    return x ** 2
```

```python
my_func
```

```python
np.testing.assert_equal(my_func(5), 25)
```

```python
np_func = np.frompyfunc(my_func, 1, 1)
np_func
```

```python
a = np.arange(10)
np_func(a)
```

```python
b = np.outer(a, a)
b
```
