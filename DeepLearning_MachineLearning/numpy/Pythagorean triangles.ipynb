```python
def is_integer(number):
    return np.equal(np.mod(number, 1), 0)
```

```python
np_func = np.frompyfunc(is_integer, 1, 1)
```

```python
numbers_range = 9
```

```python
frist_array = np.arange(numbers_range) + 1
second_array = np.arange(numbers_range) + 1
frist_array, second_array
```

```python
result = np.add.outer(frist_array ** 2, second_array ** 2)
result
```

```python
final = np.sqrt(result)
final
```

```python
np_func(final)
```
