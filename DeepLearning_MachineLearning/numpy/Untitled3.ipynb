```python
import numpy as np
```

```python
a = np.random.randint(1, 100, size=(2, 3, 3))
a
```

```python
unique_array = np.unique(a)
unique_array
```

```python
a[a > 30]
```

```python
greater_than_30 = a > 30
greater_than_30
```

```python
even_elements = a % 2 == 0
even_elements
```

```python
final_mask = np.logical_and(greater_than_30, even_elements)
final_mask
```

```python
a[final_mask]
```
