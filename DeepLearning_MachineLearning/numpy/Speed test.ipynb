```python
import numpy as np
import datetime
```

```python
a = np.random.randint(10, 99, size=1000)
b = np.random.randint(10, 99, size=1000)
```

```python
python_start = datetime.datetime.now()
python_sum = 0
for i, j in zip(a, b):
    python_sum += i*j
python_end = datetime.datetime.now()
python_result = python_end - python_start
python_result, python_sum
```

```python
np_start = datetime.datetime.now()
np_sum = np.dot(a, b)
np_end = datetime.datetime.now()
np_result = np_end - np_start
np_result, np_sum
```

```python
python_result / np_result
```
