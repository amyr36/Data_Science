```python
import numpy as np
import pandas as pd
```

```python
sales_number = np.random.randint(5, 25, size=(7, 3))
sales_number
```

```python
weekly_sales = pd.DataFrame(sales_number,
                            index=["saturday", "sunday", "monday", "thuesday", "wednseday", "thursday", "friday"],
                            columns=["pizza", "sandwich", "burger"])
weekly_sales
```

```python
prices = np.array([25, 15, 20])
prices.shape = (1, 3)
prices.shape
```

```python
foods_sale = pd.DataFrame(prices,
                            index=["price"],
                            columns=["pizza", "sandwich", "burger"])
foods_sale
```

```python
weekly_sales.T.shape
```

```python
total_sale = foods_sale.dot(weekly_sales.T)
total_sale
```

```python
weekly_sales["total"] = total_sale.T
weekly_sales
```
