```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name': ["amir", "iran", "mina", 2564],
    'age': [25, 12, 67, 'kk'],
    'favorit poem': ["akhvan", 1111, "ebtehag", "sepehry"]
})

df.to_csv("missing_data.csv")
```

```python
mising_values = ["kk", 2564, 1111]
df = pd.read_csv("missing_data.csv", na_values = mising_values)
df
```

```python
df.dropna(axis=0, how = "all")
```

```python
df["age"].fillna(df["age"].mean())
```

```python
df["age"].fillna(10101)
```

```python
df["age"].fillna(method = "ffill") # bfill
```

```python
df["age"].interpolate(method = "linear")
```
