```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('weather.csv')
weather = weather.rename(columns = {"Date.Full": "datetime"})
weather["datetime"] = pd.to_datetime(weather["datetime"])

min = weather["datetime"].min()
max = weather["datetime"].max()
max - min

weather["Date.Month"] - weather["datetime"].dt.month

weather.head(3)
```

```python
fig, axs = plt.subplots(figsize = (12, 4))
weather.groupby([weather["datetime"].dt.day])["Data.Wind.Speed"].mean().plot(kind = "bar", ax = axs)
plt.xlabel("days of month")
plt.ylabel("wind speed(m/s)")
```
