```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hospital_dataset_extended.csv')
df = df.drop(columns=['PatientID', 'RoomNumber'])
df['AdmissionDate'] = pd.to_datetime(df['AdmissionDate'])
df['DischargeDate'] = pd.to_datetime(df['DischargeDate'])
df['StayDuration'] = (df['DischargeDate'] - df['AdmissionDate']).dt.days
```

```python
avg_age = df['Age'].mean()
print(f'Average age of patients: {avg_age}')
```

```python
common_diagnosis = df['Diagnosis'].value_counts().head()
print(f'Five common diagnosis: {common_diagnosis}')
```

```python
cost_by_department = df.groupby('Department')['TreatmentCost'].mean()
print(f'Average cost of treatment by department: {cost_by_department}')
```

```python
department_counts = df['Department'].value_counts()
department_counts.plot(kind='bar', title='Patients in each department')
plt.xlabel('Departments')
plt.ylabel('Number of patients')
plt.show()
```
