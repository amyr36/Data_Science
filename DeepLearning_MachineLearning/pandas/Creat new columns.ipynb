```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
titanic['new_col'] = titanic['Age'] / titanic['Fare']
```

```python
titanic_renamed = titanic.rename(
    columns = {
        'Ticket': 'Bilit'
    }
)
```

```python
titanic[["Age", "Fare"]].describe()

titanic.agg(
    {
        "Age": ["min", "max", "median"],
        "Fare": ["min", "max", "median"]
    }
)
```

```python
titanic[['Age', 'Sex']].groupby("Sex").mean()
titanic.groupby('Sex')['Age'].mean()

print(titanic.groupby('Pclass')['Pclass'].count()) # dropna = True
titanic['Pclass'].value_counts()
```
