```python
import pandas as pd
```

```python
df = pd.read_csv('IMDB_movies.csv', usecols=['title', 'date_published', 'genre', 'country', 'director', 'avg_vote', 'worlwide_gross_income']).tail(500)
```

```python
df = df.rename(columns={'worlwide_gross_income': 'gross'})
df['gross'] = df['gross'].str.replace('$', '').astype(float)
print(df.isnull().sum())
df = df.dropna(subset=['gross'])
df = df.reset_index(drop=True)
df
```

```python
top_avg_vote = df.sort_values('avg_vote', ascending=False)
top_avg_vote[['title', 'genre', 'avg_vote']]
```

```python
top_gross = df.sort_values('gross', ascending=False)
top_gross[['title', 'genre','country', 'gross']]
```

```python
director_stats = df.groupby('director').agg({
    'avg_vote': 'mean',
    'gross': 'mean',
    'title': 'count'
}).rename(columns={'title': 'title_count'})
director_stats.sort_values('gross', ascending=False).head()
```
