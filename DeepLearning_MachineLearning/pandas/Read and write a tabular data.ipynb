```python
import pandas as pd
import numpy as np
```

```python
weight = np.random.uniform(60, 100, size=(150))
rounded_weight = np.around(weight, decimals=1)

height = np.random.uniform(1.5, 2.1, size=(150))
rounded_height = np.around(height, decimals=1)

ages = np.random.randint(18, 39, size=150)

BMI = rounded_weight / rounded_height**2
rounded_BMI = np.around(BMI, decimals=1)

BMI_data = pd.DataFrame({
    "weight": rounded_weight,
    "height": rounded_height,
    "age": ages,
    "BMI": rounded_BMI
})
BMI_data
BMI_data.to_csv("BMI_data.csv")
```
